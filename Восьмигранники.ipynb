```python
from itertools import product

# Генерируем все возможные комбинации чисел от 3 до 7 для восьмизначных последовательностей
sequences = [list(seq) for seq in product([3, 4, 5, 6, 7], repeat=8)]

# Приводим все последовательности к одному стандарту с сортировкой
unique_sequences = []
for seq in sequences:
    sorted_seq = sorted(seq)
    if sorted_seq not in unique_sequences:
        unique_sequences.append(sorted_seq)

print(unique_sequences)
```

```python
mass = []
k = 1
for i in range(len(unique_sequences)):
    summ = 0
    for j in range(8):
        summ += unique_sequences[i][j]
    if summ % 2 == 0:
        mass.append(unique_sequences[i].copy())
        k += 1
mass
```
